```python
df = df[~df["Description"].str.contains("NYC", case=False, na=False)]
```

```python
# Ensure 'Cost' column is numeric
df["Cost"] = pd.to_numeric(df["Cost"], errors="coerce")

# Ensure 'Date' column is in datetime format (if necessary)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Now perform the calculation
average_cost_per_date = df["Cost"].sum() / len(df["Date"].unique())

print(average_cost_per_date/4)
```

```python
# Ensure 'Date' is in datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Convert 'Cost' to numeric
df["Cost"] = pd.to_numeric(df["Cost"], errors="coerce")

# Extract month
df["Month"] = df["Date"].dt.to_period("M")

# Calculate month-wise total expenditure
month_wise_expenditure = df.groupby("Month")["Cost"].sum()

# Divide by 6 for per person cost
month_wise_per_person = month_wise_expenditure / 4

# Print result
print("Month-wise per person cost:\n", month_wise_per_person.round(2))
```
